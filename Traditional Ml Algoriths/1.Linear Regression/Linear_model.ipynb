```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
observations=1000
xs=np.random.uniform(low=-10,high=10,size=(observations,1))
zs=np.random.uniform(-10,10,(observations,1))

input=np.column_stack((xs,zs))
print(input.shape)
```

```
noise=np.random.uniform(-1,1,(observations,1))

targets=2*xs+3*zs+5+noise
print(targets.shape)
```

```
targets=targets.reshape(observations,)

fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('targets')
ax.view_init(azim=1000)
plt.show()
targets=targets.reshape(observations,1)
```

```
init_range=0.1

weights=np.random.uniform(-init_range,init_range,size=(2,1))
baise=np.random.uniform(-init_range,init_range,size=1)
print(weights)
print(baise)
```

```
learning_rate=0.02

for i in range(1000):
  outputs=np.dot(input,weights)+baise

  delta=outputs-targets
  loss=np.sum(delta**2)/2/observations
  print(loss)

  delta_scaled=delta/observations
  weights=weights-learning_rate*np.dot(input.T,delta_scaled)
  baise=baise-learning_rate*np.sum(delta_scaled)
```

```
print(weights,baise)
```

```
plt.plot(targets,outputs)
plt.xlabel('Targets')
plt.ylabel('outputs')
plt.title('Summary')
```
